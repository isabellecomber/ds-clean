# Example using our library
This example uses the library created for clustering.

```python
# Import our library and anything we need in this notebook
import lfsklearn as lf
import pandas as pd
%matplotlib inline
```

```python
# Let's load some data relating to nation happiness
happy = pd.read_csv("happy2017.csv")
happy.head()
```

```python
# Select columns we want to cluster on
happyCols = ['Economy..GDP.per.Capita.','Family','Health..Life.Expectancy.','Freedom', 
             'Generosity','Trust..Government.Corruption.', 'Dystopia.Residual']
```

```python
# Run the clustering using our new function
happyToCluster = happy[happyCols]
clusters = lf.Cluster(happyToCluster, 3)
```

```python
# Add the clusters to the original data
happy["clusters"] = clusters
```

```python
# Display scatter plots with colours for our clusters
for col in happyCols:
    lf.ClusterScatter(happy, 'Happiness.Score', col, 'clusters')
```

```python
# Use PCA to map the clusters to a 2D space
lf.PCAPlot(happyToCluster, 3)
```
